chính của các khách sạn nhằm tăng hiệu quả tạo doanh thu và để chúng tôi đưa ra lời
khuyên kinh doanh thấu đáo nhằm giải quyết vấn đề này.

DATA ANALYSIS AND VISUALIZATION

```python
cancelled_perc = df['is_canceled'].value_counts(normalize = True)
```

```python
#Số lượng phòng bị hủy và không bị hủy
plt.figure(figsize = (6,4))
plt.title("Reservation status count", size = 15)
plt.bar(["Not canceled", "Canceled"],df['is_canceled'].value_counts(), edgecolor = "black", width = 0.7 )
plt.savefig("pnj")
```

```python
df
```

```python
#Số lượng phòng bị hủy và không hủy ở Resort và city hotel
plt.figure(figsize = (8,4))
sns.countplot(x= "hotel", hue = "is_canceled", data = df, palette = "Blues")
plt.ylabel("Number of reservation")
plt.legend(["Not canceled", "canceled"])
plt.title("Reservation status in different hotels", size = 15)
plt.savefig("2pnj")
```

```python
resort_hotel = df[df["hotel"]=="Resort Hotel"]
resort_hotel["is_canceled"].value_counts(normalize = True)
```

```python
city_hotel = df[df["hotel"]=="City Hotel"]
city_hotel["is_canceled"].value_counts(normalize = True)
```

```python
resort_hotel = resort_hotel.groupby(by="reservation_status_date")[["adr"]].mean()
city_hotel = city_hotel.groupby(by="reservation_status_date")[["adr"]].mean()
```

```python
resort_hotel
```

```python
plt.figure(figsize = (20,8))
plt.title("Average Daily Rate in City and Resort Hotel", fontsize = 20)
plt.plot(resort_hotel.index, resort_hotel["adr"], label = "Resort Hotel")
plt.plot(city_hotel.index, city_hotel["adr"], label = "City Hotel")
plt.legend(fontsize = 20)
plt.show
plt.savefig("3pnj")
```

```python
#Số lượng hủy phòng mỗi tháng
df["month"] = df["reservation_status_date"].dt.month
plt.figure(figsize = (16,8))
sns.countplot(x = "month", hue = "is_canceled", data = df, palette = "bright" )
plt.title("Reservation status per month", size = 20)
plt.xlabel("month")
plt.ylabel("number of reservations")
plt.legend(["not canceled","canceled"])
plt.savefig("4pnj")
```

```python
#Tỷ lệ hủy ở các quốc gia
canceled_data = df[df["is_canceled"] ==1]
top_10_country = canceled_data["country"].value_counts()[:10]
plt.figure(figsize = (8,8))
plt.title("Top 10 country reservation canceled")
plt.pie(top_10_country, autopct ="%.2f",labels = top_10_country.index)
plt.savefig("7pnj")
```

```python
canceled_df_adr = canceled_data.groupby("reservation_status_date")[["adr"]].mean()
canceled_df_adr.reset_index(inplace = True)
canceled_df_adr.sort_values("reservation_status_date",inplace = True)
canceled_df_adr

not_canceled_data = df[df["is_canceled"]==0]
not_canceled_df_adr = not_canceled_data.groupby("reservation_status_date")[["adr"]].mean()
not_canceled_df_adr.reset_index(inplace = True)
not_canceled_df_adr.sort_values("reservation_status_date",inplace = True)

plt.figure(figsize = (20,9))
plt.title("Average Daily Rate")
plt.plot(not_canceled_df_adr["reservation_status_date"], not_canceled_df_adr["adr"], label = "not canceled")
plt.plot(canceled_df_adr["reservation_status_date"],canceled_df_adr["adr"], label = "canceled")
plt.legend()
plt.xlabel("reservation_status_date")
plt.ylabel("adr")
plt.savefig("9pnj")
```
